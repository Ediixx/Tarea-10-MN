1. Realice las siguiente multiplicaciones matriz-matriz:



Literal a,b,c y d:

En esta ocasión, vamos a realizar las multiplicaciones de matrices para todos los literales (a, b, c y d) en una sola ejecución. Para ello, hemos almacenado los pares de matrices en una lista llamada matrices. Posteriormente, utilizamos un bucle for para recorrer todos los pares de matrices, verificando si se pueden multiplicar. Si es el caso, se realiza la operación de multiplicación y se imprime la respuesta. Si no se pueden multiplicar, se captura la excepción y se imprime un mensaje de error.

```python
import numpy as np

def multiplicacion_matrices(matriz_a, matriz_b):
    # Verificar si las matrices pueden multiplicarse
    if len(matriz_a[0]) != len(matriz_b):
        raise ValueError("El número de columnas de la primera matriz debe ser igual al número de filas de la segunda matriz.")
    
    respuesta = np.dot(matriz_a, matriz_b)
    return respuesta

literales_mm = [
    ([[2, -3], [3, -1]], [[1, 5], [2, 0]]),  # Literal a
    ([[2, -3], [3, -1]], [[1, 5, -4], [-3, 2, 0]]),  # Literal b
    ([[2, -3, 1], [4, 3, 0], [5, 2, -4]], [[0, 1, -2], [1, 0, -1], [2, 3, -2]]),  # Literal c
    ([[2, 1, 2], [-2, 3, 0], [2, -1, 3]], [[1, -2], [-4, 1], [0, 2]])  # Literal d
]

# Realizar las multiplicaciones e imprimir los resultados
for i, (matriz_a, matriz_b) in enumerate(literales_mm):
    try:
        resultado = multiplicacion_matrices(matriz_a, matriz_b)
        print(f"Resultado de la multiplicación de matrices del literal {chr(97 + i)}:")
        print(np.array(resultado))
    except ValueError as e:
        print(f"Error en el literal {chr(97 + i)}: {e}")
```

2. Determine cuáles de las siguientes matrices son no singulares y calcule la inversa de esas matrices:




Literal a,b,c,d

En esta ocasión, vamos a determinar cuáles de las matrices son no singulares (invertibles) y calcular la inversa de esas matrices en una sola ejecución. Para ello, hemos almacenado las matrices en una lista llamada literales_ns_i. Posteriormente, utilizamos un bucle for para recorrer todas las matrices, verificando si son no singulares. Si es el caso, se calcula la inversa de la matriz y se imprime la respuesta. Si la matriz es singular (no invertible), se imprime un mensaje indicando que no tiene inversa.

```python
import numpy as np

def no_es_singular(matriz):
    # Verificar si la matriz es cuadrada
    if len(matriz) != len(matriz[0]):
        return False
    # Calcular el determinante
    determinante = np.linalg.det(matriz)
    # Una matriz es no singular si su determinante es diferente de cero
    return determinante != 0

def matriz_inversa(matriz):
    return np.linalg.inv(matriz)

literales_ns_i = [
    [[4, 2, 6], [3, 0, 7], [-2, -1, -3]],  # Literal a
    [[1, 2, 0], [2, 1, -1], [3, 1, 1]],  # Literal b
    [[1, 1, -1, 1], [1, 2, -4, -2], [2, 1, 1, 5], [-1, 0, -2, -4]],  # Literal c
    [[4, 0, 0, 0], [6, 7, 0, 0], [9, 11, 1, 0], [5, 4, 1, 1]]  # Literal d
]

# Verificar e imprimir las matrices no singulares y sus inversas
for i, matriz in enumerate(literales_ns_i):
    if no_es_singular(matriz):
        inversa = matriz_inversa(matriz)
        print(f"La matriz del literal {chr(97 + i)} es no singular.")
        print("Matriz inversa:")
        print(np.array(inversa))
    else:
        print(f"La matriz del literal {chr(97 + i)} es singular y no tiene inversa.")
```

3.  Resuelva los sistemas lineales 4 x 4 que tienen la misma matriz de coeficientes:




Literal a y b

En esta ocasión, vamos a resolver dos sistemas de ecuaciones lineales 4x4 que tienen la misma matriz de coeficientes en una sola ejecución. Para ello, hemos almacenado la matriz de coeficientes en una variable llamada A y los vectores de términos independientes en una lista llamada b_list. Posteriormente, utilizamos un bucle for para recorrer todos los vectores de términos independientes, resolviendo cada sistema lineal. Si el sistema se puede resolver, se imprime la solución con los valores de (x_1), (x_2), (x_3) y (x_4) redondeados a 5 decimales. Si no se puede resolver, se captura la excepción y se imprime un mensaje de error.

```python
import numpy as np

# Matriz de coeficientes común (A)
A = np.array([
    [1, -1, 2, -1],
    [1, 0, -1, 1],
    [2, 1, 3, -4],
    [-1, -1, 1, -1]
])

# Lista de vectores de términos independientes
b_list = [
    np.array([6, 4, -2, 5]),   # Primer sistema
    np.array([1, 1, 2, -1])    # Segundo sistema
]

# Resolver los sistemas e imprimir los resultados
for i, b in enumerate(b_list):
    try:
        # Resolver el sistema Ax = b
        x = np.linalg.solve(A, b)
        x = np.round(x, 5)
        print(f"Solución del sistema {i + 1}:")
        for j in range(len(x)):
            print(f"x{j + 1} = {x[j]}")
    except np.linalg.LinAlgError as e:
        print(f"Error al resolver el sistema {i + 1}: {e}")
```

4.  Encuentre los valores de A que hacen que la siguiente matriz sea singular.




En esta ocasión, vamos a determinar los valores del parámetro (\alpha) que hacen que el determinante de una matriz específica sea igual a cero. Para ello, hemos definido la matriz (A) utilizando el símbolo (alpha) y hemos calculado su determinante. Posteriormente, resolvemos la ecuación determinante = 0 para (\alpha). Si se encuentran soluciones, se redondean a tres decimales y se imprimen. Si no se encuentran soluciones, se captura la excepción y se imprime un mensaje de error.

```python
import sympy as sp

# Redefinir el símbolo alpha
alpha = sp.Symbol('alpha')

# Definir la matriz A
A = sp.Matrix([[1, -1, alpha],[2, 2, 1],[0, alpha, -3/2]])

# Calcular el determinante
determinante = A.det()

# Resolver la ecuación determinante = 0 para alpha
soluciones_alpha = sp.solve(determinante, alpha)

# Redondear las soluciones a tres decimales
soluciones_redondeadas = [float(sol.evalf(3)) for sol in soluciones_alpha]

# Mostrar los resultados
print(f"α₁ = {soluciones_redondeadas[0]}")
print(f"α₂ = {soluciones_redondeadas[1]}")
```

7. Modifique el algoritmo de eliminación gaussiana de tal forma que se pueda 
utilizar para resolver un sistema lineal usando la descomposición LU y, a continuación, 
resuelva los siguientes sistemas lineales




En esta ocasión, vamos a resolver varios sistemas de ecuaciones lineales utilizando la descomposición LU (LU decomposition) y la eliminación gaussiana en una sola ejecución. Para ello, hemos definido las matrices de coeficientes y los vectores de términos independientes para cada sistema. Posteriormente, utilizamos un bucle for para recorrer todos los sistemas, realizando la descomposición LU de cada matriz y resolviendo el sistema lineal correspondiente. Si el sistema se puede resolver, se imprimen las matrices (L) y (U) y la solución del sistema. Si no se puede resolver, se captura la excepción y se imprime un mensaje de error.

```python
import numpy as np

def eliminacion_gaussiana_lu(A):
    n = A.shape[0]
    L = np.eye(n, dtype=np.float64)  # Matriz identidad para L
    U = A.astype(np.float64)  # Asegurar que U sea de tipo float

    for i in range(n):
        for j in range(i + 1, n):
            if U[i, i] == 0:
                raise ValueError("El pivote es cero. La matriz puede ser singular.")
            # Factor multiplicador
            factor = U[j, i] / U[i, i]
            L[j, i] = factor
            # Actualización de la fila
            U[j, i:] -= factor * U[i, i:]
    
    return L, U

def resolver_sistema_lu(L, U, b):
    # Sustitución hacia adelante para resolver Ly = b
    n = L.shape[0]
    y = np.zeros_like(b, dtype=np.float64)

    for i in range(n):
        y[i] = b[i] - np.dot(L[i, :i], y[:i])

    # Sustitución hacia atrás para resolver Ux = y
    x = np.zeros_like(b, dtype=np.float64)

    for i in range(n - 1, -1, -1):
        x[i] = (y[i] - np.dot(U[i, i + 1:], x[i + 1:])) / U[i, i]

    return x

def funcion_principal():
    sistemas = [(np.array([[2, -1, 1], [3, 3, 9], [3, 3, 5]], dtype=np.float64), np.array([-1, 0, 4], dtype=np.float64)), # Sistema a
    
        (np.array([[1.012, -2.132, 3.104], [-2.132, 4.096, -7.013], [3.104, -7.013, 0.014]]),np.array([1.984, -5.049, -3.895])),# Sistema b

        (np.array([[2, 0, 0, 0], [1, 1.5, 0, 0], [0, -3, 0.5, 0], [2, -2, 1, 1]]), np.array([3, 4.5, -6.6, 0.8])), # Sistema c
        
        (np.array([[2.1756, 4.0231, -2.1732, 5.1967],[-4.0231, 6.0000, 0, 1.1973],[-1.0000, -5.2107, 1.1111, 0],[6.0235, 7.0000, 0, -4.1561]]),np.array([17.102, -6.1593, 3.0004, 0.0000]))    # Sistema d
    ]

    for idx, (A, b) in enumerate(sistemas):
        print(f"\nResolviendo Literal {chr(97 + idx)}:")
        try:
            L, U = eliminacion_gaussiana_lu(A)
            print("Matriz L:")
            print(L)
            print("Matriz U:")
            print(U)
            x = resolver_sistema_lu(L, U, b)
            print(f"Solución: {x}")
        except ValueError as e:
            print(f"Error: {e}")

funcion_principal()
```
